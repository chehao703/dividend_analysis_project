<h3>進模型前處理空值<h3>

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv(r'C:\henry_Chang\final_project\model\data\ML_total_industry_lastest_v3.csv')
```

```python
# 刪除包含空值資料
df1 = df1.dropna(subset = ['earning_per_share','stocks'])
```

```python
# 刪除多數空值欄位
df1 = df1.drop(['other_comprehensive_income','total_comprehensive_income','parent_comprehensive_income',
                'control_comprehensive_income','non_control_comprehensive_income'],axis = 1)
```

```python
# 欄位空值補0
values={'jointly_control_PL':0,'non_control_PL':0,'control_interest':0,'non_control_interest':0}
df1=df1.fillna(values)
```

```python
# 補上 2012年 6113 的 total_treasury_shares (庫藏股數（母持及子持母）) 為5560
df1['total_treasury_shares'][19581] = 5560
```

```python
# 確認無空值
df1.isnull().sum()
```

```python
# 確認無重複值
df1.duplicated().sum()
```

```python
df1['all_dividend']=0
```

```python
def add_dividend(row):
    row['all_dividend'] = row['cash_dividend']+row['stock_dividend']
    return row
```

```python
df1=df1.apply(add_dividend,axis=1)
```
